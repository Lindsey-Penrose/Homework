#### Question 3 ####
```SQL
select f.facilityID, d.typeFacility, count(*) as numCharges
from factcharge f
LEFT join dimFacility d
on f.facilityId = d.Facilitykey
group by d.FacilityKey
order by numCharges desc
```

    A primary key is a key within one table that cannot have duplicate values. A foreign key is a key from another table that you use to link one table to another; it's the one that shares similar data to the primary table.
