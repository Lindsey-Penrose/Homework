```python
# 2. Chained "if" statement

age = input("How old are you?")
if int(age) >= 25:
    print("Renting a car is more affordable.")
elif int(age) > 18 and int(age) < 25:
    print("Renting a car is very expensive.")
else:
    print("You can't drive a car!")
```

```python
# 4. Writing a nested loop
couture = ["trousers", "blouse", "bandana", 
        "cumber band", "blazer", "vest", "french beret", 
        "scarf", "stole"]
panettone = {"cerise", "fuchsia", "aqua", "maple", "auburn", 
             "burnt sienna", "gunmetal blue", "Dark Sapphire"}

for item in couture:
    for var in panettone:
        print(item, var)

```
